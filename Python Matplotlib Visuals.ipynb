The data for a bar plot is different from the data we've seen up until now. For line charts, we were using numeric data on both the x and y axes. For a bar plot, we need our x-axis data to be a set of labels for the bars, and the y-axis data should be a numeric value that corresponds to the height of each bar.

```python
pl1=df.Embarked.value_counts()
pl1
```

## Histograms

A histogram is a graphical representation of the distribution of a dataset. It divides the data into intervals or "bins" and shows the frequency or count of data points that fall into each bin. Histograms are commonly used to visualize the underlying frequency distribution of a continuous variable.

```python
age=df['Age']
age
```

```python
plt.hist(age,color='skyblue',bins=10,edgecolor='black')
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('Passengers')
plt.title('Distribution of Passenger Ages');
```

**From this distribution, we can see that most of the passengers were in the age range of about 20 to 40. 
We can also see that there were many young children aboard, and relatively few passengers over the age of 70.**

**Note that this illustrates the total count (y-axis) of the ages (x-axis) of the passengers separated into 10 bins. 
We can show a probability density of ages in each bin instead of the total count by changing to a density chart**

```python
plt.hist(age,color='skyblue',bins=10,edgecolor='black',density=True)
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('Density Probability')
plt.title('Distribution of Passenger Ages')
```

## PIE CHARTS
To make a pie chart, use Matplotlib's matplotlib.pyplot.pie function, which expects a list representing the sizes of each slice. We can use the count of male and female passengers as an example.

```python
x=df['Sex'].value_counts()
x
```

```python
plt.pie(x,labels=['Male','Female']);
```

```python
#That's better, but what if we want the first slice to start at the 12 o'clock position, which is standard? The default seems to be starting at 3 o'clock. 
#We can do that by specifying a startangle.
```

```python
plt.pie(x,labels=['Male','Female'],startangle=90);
```

```python
#You can separate one or more slices from the rest of the chart using the explode parameter.
plt.pie(x,labels=['Male','Female'],startangle=90,explode=[0.07,0]);
```

```python
#Finally, we can display the percentage that each slice represents using the autopct parameter by setting it to a format specifier.
plt.pie(x,labels=['Male','Female'],startangle=90,explode=[0.07,0],autopct='%.1f%%');
```

## Scatter plots
A scatter plot is a two-dimensional data visualization technique that represents individual data points as dots or markers on a Cartesian plane. Each dot on the plot corresponds to a single data point, with one axis (usually the x-axis) representing one variable and the other axis (usually the y-axis) representing another variable. Scatter plots are particularly useful for visualizing the relationship or association between two continuous variables.

```python
df.head()
```

```python
age=df['Age']
fare=df['Fare']
pclass=df['Pclass']
plt.figure(figsize=(12,6))
plt.scatter(age,fare,pclass)
```

```python
age=df['Age']
fare=df['Fare']
pclass=df['Pclass']
plt.figure(figsize=(8,6))
plt.scatter(age,fare,pclass,cmap='RdYlGn')
plt.colorbar(ticks=[1, 2, 3])
plt.xlabel('Age')
plt.ylabel('Fare');
```

## BOX PLOTS
You can create box plots (box-and-whisker plots) in Matplotlab using the matplotlib.pyplot.boxplot function. Box plots are important in statistical analysis because they let you see the distrubution of a data set at a glance, including outliers. 
Let's look at the age of Titanic passengers again, this time in a box plot.

```python
age=df['Age'].dropna()
plt.figure(figsize=(7,7))
plt.boxplot(age,labels=['Age']);
```

**We can pick a few key datapoints off the boxplot at a glance. The median age is just under 30, with half of the ages in the 20 to 40 range. 
The entire distribution of ages is from 0 to 80 years, with a few data points at the high end of the scale considered outliers.**

## 3D plots
Earlier we showed how color and marker size can be used as a third dimension in Matplotlib, but you can also plot that data on a third axis by setting the projection to 3d when calling the axes function. That function returns an axis object that we use to create a scatterplot.

```python
age = df['Age']
fare = df['Fare']
pclass = df['Pclass']

fig = plt.figure(figsize=(8, 8))

ax = plt.axes(projection='3d')
ax.scatter(xs=age, ys=fare, zs=pclass, c=pclass, cmap='brg')

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Class');
```
